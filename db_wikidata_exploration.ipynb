```python
# Importing Packages

import csv
import pandas as pd
import numpy as np
import re
```

```python
# Importing CSV files (of the queries) & Concanating all query files

q_1 = pd.read_csv (r'query_developer_publisher_designed_composer.csv')
q_2 = pd.read_csv (r'query_fiction_follows_followed_basedon.csv')
q_3 = pd.read_csv (r'query_gamemode.csv')
q_4 = pd.read_csv (r'query_origin_pubdate_genre.csv')
q_5 = pd.read_csv (r'query_plattform.csv')
q_6 = pd.read_csv (r'query_web_esrb_cero_pegi.csv')

frames=[q_1,q_2,q_3,q_4,q_5,q_6]
df_wikidata = pd.concat(frames)
groups = df_wikidata.groupby(['videogames','videogamesLabel']).agg(lambda x: list(x)) 

groups_test = groups
```

```python
# Removing the NaN string from every list in the dataframe

groups_test['developerLabel'] = [[i for i in L if str(i) not in remove_list] for L in groups_test['developerLabel']]
groups_test['publisherLabel'] = [[i for i in L if str(i) not in remove_list] for L in groups_test['publisherLabel']]
groups_test['designedByLabel'] = [[i for i in L if str(i) not in remove_list] for L in groups_test['designedByLabel']]
groups_test['composerLabel'] = [[i for i in L if str(i) not in remove_list] for L in groups_test['composerLabel']]
groups_test['fictionalUniverseLabel'] = [[i for i in L if str(i) not in remove_list] for L in groups_test['fictionalUniverseLabel']]
groups_test['followsLabel'] = [[i for i in L if str(i) not in remove_list] for L in groups_test['followsLabel']]
groups_test['followedByLabel'] = [[i for i in L if str(i) not in remove_list] for L in groups_test['followedByLabel']]
groups_test['basedOnLabel'] = [[i for i in L if str(i) not in remove_list] for L in groups_test['basedOnLabel']]
groups_test['gameModeLabel'] = [[i for i in L if str(i) not in remove_list] for L in groups_test['gameModeLabel']]
groups_test['countryOriginLabel'] = [[i for i in L if str(i) not in remove_list] for L in groups_test['countryOriginLabel']]
groups_test['publicationDate'] = [[i for i in L if str(i) not in remove_list] for L in groups_test['publicationDate']]
groups_test['genreLabel'] = [[i for i in L if str(i) not in remove_list] for L in groups_test['genreLabel']]
groups_test['plattformLabel'] = [[i for i in L if str(i) not in remove_list] for L in groups_test['plattformLabel']]
groups_test['WebsiteLabel'] = [[i for i in L if str(i) not in remove_list] for L in groups_test['WebsiteLabel']]
groups_test['ESRBratingLabel'] = [[i for i in L if str(i) not in remove_list] for L in groups_test['ESRBratingLabel']]
groups_test['CEROratingLabel'] = [[i for i in L if str(i) not in remove_list] for L in groups_test['CEROratingLabel']]
groups_test['PEGIratingLabel'] = [[i for i in L if str(i) not in remove_list] for L in groups_test['PEGIratingLabel']]
```

```python
# replace empty lists with NaN & count number of empty attributes per column

groups_test.to_csv('help.csv')
helper = pd.read_csv (r'help.csv', na_values='[]')
```

```python
# Count number of NaNs per Column
helper.isnull().sum(axis = 0)
```

```python
# Building Subsets of all videogames based on missing values
# 1 = rows with no NaN = 0
print('Number of videogames in Wikidata = '+str(len(helper)))
print('---------------------------------------')

subset_1 = helper.dropna()
print('Number rows with no NaN = '+str(len(subset_1)))
# 2 = drop all rows which contain more than 3 NaN Values = 153
subset_2 = helper.dropna(thresh=helper.shape[1]-3, axis=0)
print('Number rows with at most 3 NaN = '+str(len(subset_2)))
# 2 = drop all rows which contain more than 7 NaN Values = 29xx
subset_3 = helper.dropna(thresh=helper.shape[1]-7, axis=0)
print('Number rows with at most 7 NaN = '+str(len(subset_3)))
print('---------------------------------------')

subset_4 = helper.dropna(how='any', subset=['plattformLabel','publicationDate'])
print('Number rows with when plattform & pub Date are not missing = '+str(len(subset_4)))
```

```python
# List of videogames included in this dataframe

list_videogames = list(set(helper['videogamesLabel']))
print("Number of videogames total: "+ str(len(list_videogames)))

# delete all with string (start with Q and then just numbers) = video games we dont have the name for
for i in list_videogames:
    if re.match('^[Q][\d]+', i):
        list_videogames.remove(i)
        

print('Number of videogames where we know the name: '+str(len(list_videogames)))
```

```python
# save to csv_file
df = pd.DataFrame(list_videogames)
df.to_csv("./videogames_list_wiki.csv", sep=',',index=True)
```
