```python
# Importing Packages
import csv
import pandas as pd
import numpy as np
```

```python
# Load data
wikidata = pd.read_excel(r'videogames_wiki_filtered.xlsx')
sales = pd.read_csv(r'd1_vgsales.csv')
```

```python
# WIKIDATA
# Drop unnecessary columns, 
# everything lowercase, 
# converts string into lists and only keeps unique values in list
# adds rank with continious numbers from 16601 for gold standard

wikidata = wikidata.drop(["fictionalUniverseLabel", "basedOnLabel", "Unnamed: 0"], axis=1)
wikidata = wikidata.applymap(lambda s:s.lower() if type(s) == str else s)
wikidata = wikidata.applymap(lambda x: x.strip('[]').lstrip(' ') if type(x) == str else x)
wikidata = wikidata.applymap(lambda x: x.replace('\'','') if type(x) == str else x)
wikidata = wikidata.applymap(lambda s:s.split(", ") if s != 'videogamesLabel' and type(s)==str else s)
wikidata = wikidata.applymap(lambda s:s.lstrip(' ') if type(s) == str else s)
wikidata = wikidata.applymap(lambda s:list(set(s)) if type(s) == list else s)
wikidata['videogamesLabel'] = wikidata['videogamesLabel'].astype(str)
wikidata['videogamesLabel'] = wikidata['videogamesLabel'].map(lambda x: x.strip('[]').replace('\'',''))
wikidata.insert(0,"Rank",range(16601, 16601+len(wikidata)))
wikidata
```

```python
# SALES
# everything lowercase

sales = sales.applymap(lambda s:s.lower() if type(s) == str else s)
sales
```

```python
# save to csv_file
wikidata.to_csv(".//wiki_preprocessed.csv", sep=',',index=False)
sales.to_csv(".//sales_preprocessed.csv", sep=',',index=False)
```
